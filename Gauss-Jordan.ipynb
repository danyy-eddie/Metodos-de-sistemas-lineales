MÉTODO GAUSS-JORDAN

```python
#SISTEMA DE ECUACIONES
#x + y + z = 6  
#2y + 5z = -4  
#2x + 5y - z = 27  
import numpy as np

# 1. Definir la matriz aumentada [A|B]
aug_matrix = np.array([
    [1, 1, 1, 6],
    [0, 2, 5, -4],
    [2, 5, -1, 27]
], dtype=float)

rows, cols = aug_matrix.shape

# 2. Gauss-Jordan paso a paso
for i in range(rows):
    # 2.1 Hacer el pivote igual a 1
    pivot = aug_matrix[i][i]
    if pivot != 0:
        aug_matrix[i] = aug_matrix[i] / pivot
    else:
        raise ValueError("Pivote es cero, requiere intercambio de filas.")

    # 2.2 Hacer ceros en la columna del pivote para las demás filas
    for j in range(rows):
        if i != j:
            factor = aug_matrix[j][i]
            aug_matrix[j] = aug_matrix[j] - factor * aug_matrix[i]

# 3. Imprimir la matriz reducida
print("Matriz reducida (forma escalonada reducida):")
print(aug_matrix)

# 4. Extraer soluciones
soluciones = aug_matrix[:, -1]
print("\nSolución del sistema:")
variables = ['x', 'y', 'z']
for var, val in zip(variables, soluciones):
    print(f"{var} = {val}")
```

El sistema se representa como una matriz aumentada:
aug_matrix = np.array([
    [1, 1, 1, 6],
    [0, 2, 5, -4],
    [2, 5, -1, 27]

El método de Gauss-Jordan transforma la matriz en su forma escalonada reducida por filas (forma canónica).
Para cada fila, se divide la fila completa entre el pivote (el número en la diagonal), para que ese número se convierta en 1.
Después de hacer el pivote 1, se hacen ceros en su columna arriba y abajo, para que la variable solo exista en una ecuación.
Esto se repite para cada variable ( x, y, z), es decir, para cada columna principal.
El código hace exactamente eso:

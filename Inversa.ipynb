MÉTODO INVERSA

```python
#SISTEMA DE ECUACIONES
#2x + 3y = 8
#5x - y = 2
import numpy as np

# Matriz de coeficientes A
A = np.array([[2, 3],
              [5, -1]])

# Vector de resultados B
B = np.array([8, 2])

# Calculamos la inversa de A
A_inv = np.linalg.inv(A)

# Mostramos la inversa
print("Matriz inversa de A:")
print(A_inv)

# Solucionamos el sistema: X = A⁻¹ * B
X = np.dot(A_inv, B)

# Mostramos el resultado
print("Solución del sistema (x, y):")
print(X)
```

El sistema de ecuaciones se puede escribir de la siguiente manera:
[ 2   3 ] [ x ] = [ 8 ]  
[ 5  -1 ] [ y ] = [ 2 ]

Esto es equivalente a:
A * X = B
Dónde:
A es la matriz de coeficientes ,
X es el vector de incógnitas (en este caso [x, y]),
B es el vector de resultados .

El método para resolver el sistema de ecuaciones consiste en calcular la inversa de la matriz A, ya que:
X = A⁻¹ * B
Donde A⁻¹es la inversa de la matriz A, y al multiplicarla por el vector B, obtenemos el vector X, que contiene las soluciones de las incógnitas xy y.

Una vez que tenemos la inversa de A, podemos multiplicarla por el vector Bpara obtener las soluciones de las incógnitas xy y:
X = np.dot(A_inv, B)
Esto realiza la multiplicación de matrices: A⁻¹ * B.

```python
#SISTEMA DE ECUACIONES
#x + y + z = 6  
#2y + 5z = -4  
#2x + 5y - z = 27  
import numpy as np
from scipy.linalg import lu, solve_triangular

# Matriz de coeficientes A
A = np.array([[1, 1, 1],
              [0, 2, 5],
              [2, 5, -1]], dtype=float)

# Vector de resultados B
B = np.array([6, -4, 27], dtype=float)

# 1. Descomponemos A en P, L y U (PA = LU)
P, L, U = lu(A)

# Mostramos las matrices resultantes
print("Matriz P:")
print(P)
print("\nMatriz L:")
print(L)
print("\nMatriz U:")
print(U)

# 2. Resolver el sistema LUx = PB
# Paso 1: resolver Ly = Pb
Pb = np.dot(P, B)
y = solve_triangular(L, Pb, lower=True)

# Paso 2: resolver Ux = y
x = solve_triangular(U, y)

# Resultado
print("\nSolución del sistema (x, y, z):")
print(x)
```

MÉTODO LU

```python
Construimos la Matriz A y el Vector B
A = np.array([
    [1, 1, 1],   # Coeficientes de la ecuación 1
    [0, 2, 5],   # Coeficientes de la ecuación 2
    [2, 5, -1]   # Coeficientes de la ecuación 3
])

from scipy.linalg import lu
P, L, U = lu(A)
Esto descompone la matriz Aen tres matrices:
P Matriz de permutación (intercambios de filas para estabilidad).
L Matriz triangular inferior (con 1s en la diagonal).
U Matriz triangular superior .
    
Queremos resolver el sistema original Ax = B.
Como PA = LU, entonces también:
LUx = PB

Resolver L·y = P·B
Pb = np.dot(P, B)
y = solve_triangular(L, Pb, lower=True)
Esto resuelve una ecuación triangular inferior , que es fácil de resolver por sustitución hacia adelante.

Resolver U·x = y
x = solve_triangular(U, y)
Aquí se resuelve una ecuación triangular superior, por sustitución hacia atrás.
```
